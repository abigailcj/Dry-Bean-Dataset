```python
import numpy as np
from scipy.spatial import distance

# Assuming class_means has been calculated
class_means = df.groupby('Class').mean()

# List of class names
class_names = ['BARBUNYA', 'BOMBAY', 'CALI', 'DERMASON', 'HOROZ', 'SEKER', 'SIRA']

# Calculate and add Euclidean distances to the DataFrame
for class_name in class_names:
    distances_col_name = f'distance_to_{class_name}_mean'
    distance_formula = np.sqrt(np.sum((df.drop('Class', axis=1) - class_means.loc[class_name])**2, axis=1))
    df[distances_col_name] = distance_formula

# Display or use the new distance columns as needed
print(df.head())
```

```python
names = list(df.columns)
classes = ["BARBUNYA", "BOMBAY", "CALI", "DERMASON", "HOROZ", "SEKER", "SIRA"]

comb = list(itertools.combinations(names, 2))  # Include the "Class" column in combinations

fig, axs = plt.subplots(5, 2, figsize=(15, 20))

class_points = {class_label: None for class_label in classes}

for ax, (x_feature, y_feature) in zip(axs.flat, comb):
    for class_label, marker in zip(classes, ["o", "s", "D", "x", "+", "^", "v"]):
        points = ax.scatter(df.loc[df["Class"] == class_label, x_feature],
                            df.loc[df["Class"] == class_label, y_feature],
                            label=class_label, marker=marker)

        if class_points[class_label] is None:
            class_points[class_label] = points

    ax.set_xlabel(x_feature)
    ax.set_ylabel(y_feature)

fig.legend(handles=[class_points[class_label] for class_label in classes], labels=classes, loc="upper right")

plt.tight_layout()
plt.show()
```
